```python
import matplotlib.pyplot as plt
import numpy as np
class explo:
    def expe():
        x_axis_1=np.random.randint(10,50,4)
        myexplode = [0.3, 0, 0, 0]
        
        plt.pie(x_axis_1, explode=myexplode)
        plt.show()  
explo.expe()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
y=np.array([40,40,15,5])


plt.pie(y,labels= ["Apples","Bananas","Cherries","Dates"],explode=[0.2,0,0,0],shadow=True)
plt.legend()
plt.show()
```
